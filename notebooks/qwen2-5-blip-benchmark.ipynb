```python
import os
import sys
import subprocess
import cv2
import torch
import numpy as np
import pandas as pd
from tqdm import tqdm
from PIL import Image, ImageEnhance
from transformers import (
    AutoProcessor, 
    AutoModelForZeroShotObjectDetection,
    Qwen2_5_VLForConditionalGeneration
)
import json
import xml.etree.ElementTree as ET
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from sklearn.metrics import (
    classification_report, 
    confusion_matrix, 
    accuracy_score,
    precision_recall_fscore_support,
    precision_recall_curve,
    roc_curve,
    auc
)
import seaborn as sns
from collections import defaultdict, Counter
import warnings
warnings.filterwarnings('ignore')

# Configuration
DATASET_FOLDER = '/kaggle/input/pedestrian-behavior-on-sidewalk'
IMAGES_FOLDER = '/kaggle/input/pedestrian-behavior-on-sidewalk' 
ANNOTATIONS_FOLDER = '/kaggle/input/pedestrian-behavior-on-sidewalk'  
OUTPUT_FOLDER = '/kaggle/working/pipeline_results'
CONFIDENCE_THRESHOLD = 0.5
IOU_THRESHOLD = 0.1
CONTAINMENT_THRESHOLD = 0.3
MIN_PERSON_AREA_PERCENTAGE = 0.5
IOU_MAPPING_THRESHOLD = 0.8

# Detection classes
DETECTION_CLASSES = ["person", "bike", "skateboard", "baby stroller", "wheelchair"]

def setup_environment():
    """Install required packages"""
    print("ğŸš€ Setting up Pedestrian Behavior Dynamics Pipeline...")
    packages = [
        "torch torchvision transformers>=4.45.0", 
        "opencv-python-headless", 
        "tqdm", 
        "matplotlib", 
        "seaborn",
        "scikit-learn",
        "pandas",
        "accelerate",
        "tiktoken",
        "qwen-vl-utils"
    ]
    
    for package in packages:
        print(f"ğŸ“¦ Installing {package}...")
        result = subprocess.run(f"pip install {package}", shell=True, capture_output=True, text=True)
        if result.returncode == 0:
            print(f"âœ… {package} installed successfully")

def calculate_iou(box1, box2):
    """Calculate Intersection over Union (IoU) between two bounding boxes"""
    x1_1, y1_1, x2_1, y2_1 = box1
    x1_2, y1_2, x2_2, y2_2 = box2
    
    # Calculate intersection
    x1_i = max(x1_1, x1_2)
    y1_i = max(y1_1, y1_2)
    x2_i = min(x2_1, x2_2)
    y2_i = min(y2_1, y2_2)
    
    if x2_i <= x1_i or y2_i <= y1_i:
        return 0.0
    
    intersection = (x2_i - x1_i) * (y2_i - y1_i)
    area1 = (x2_1 - x1_1) * (y2_1 - y1_1)
    area2 = (x2_2 - x1_2) * (y2_2 - y1_2)
    union = area1 + area2 - intersection
    
    return intersection / union if union > 0 else 0.0

def calculate_containment(box1, box2):
    """Calculate what percentage of box1 is contained within box2"""
    x1_1, y1_1, x2_1, y2_1 = box1
    x1_2, y1_2, x2_2, y2_2 = box2
    
    # Calculate intersection
    x1_i = max(x1_1, x1_2)
    y1_i = max(y1_1, y1_2)
    x2_i = min(x2_1, x2_2)
    y2_i = min(y2_1, y2_2)
    
    if x2_i <= x1_i or y2_i <= y1_i:
        return 0.0
    
    intersection = (x2_i - x1_i) * (y2_i - y1_i)
    area1 = (x2_1 - x1_1) * (y2_1 - y1_1)
    
    return intersection / area1 if area1 > 0 else 0.0

def should_merge_person_nonperson(det1, det2, iou_threshold=0.1, containment_threshold=0.7):
    """Modified merging logic: Only merge person with non-person labels"""
    box1, box2 = det1['box'], det2['box']
    
    # Calculate IoU
    iou = calculate_iou(box1, box2)
    if iou > iou_threshold:
        return True, "iou"
    
    # Calculate areas
    area1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area2 = (box2[2] - box2[0]) * (box2[3] - box2[1])
    
    # Check containment for smaller objects (like skateboards in person boxes)
    if area1 < area2:  # box1 is smaller
        containment = calculate_containment(box1, box2)
        if containment > containment_threshold:
            return True, f"containment_{containment:.2f}"
    else:  # box2 is smaller
        containment = calculate_containment(box2, box1)
        if containment > containment_threshold:
            return True, f"containment_{containment:.2f}"
    
    return False, "no_merge"

def merge_person_nonperson_boxes(detections, iou_threshold=0.1, containment_threshold=0.7):
    """Merge overlapping bounding boxes but ONLY merge person with non-person"""
    if len(detections) <= 1:
        return detections
    
    merged = []
    used = set()
    
    for i, det1 in enumerate(detections):
        if i in used:
            continue
            
        overlapping = [det1]
        used.add(i)
        merge_reasons = []
        
        for j, det2 in enumerate(detections[i+1:], i+1):
            if j in used:
                continue
                
            should_merge, reason = should_merge_person_nonperson(det1, det2, iou_threshold, containment_threshold)
            
            if should_merge:
                overlapping.append(det2)
                merge_reasons.append(f"box_{j}_{reason}")
                used.add(j)
        
        if len(overlapping) == 1:
            merged.append(overlapping[0])
        else:
            # Merge overlapping detections - take the one with highest confidence
            best_det = max(overlapping, key=lambda x: x['confidence'])
            
            # Expand bounding box to encompass all overlapping boxes
            all_boxes = [det['box'] for det in overlapping]
            x1_min = min(box[0] for box in all_boxes)
            y1_min = min(box[1] for box in all_boxes)
            x2_max = max(box[2] for box in all_boxes)
            y2_max = max(box[3] for box in all_boxes)
            
            merged_det = {
                'box': [x1_min, y1_min, x2_max, y2_max],
                'confidence': best_det['confidence'],
                'merged_count': len(overlapping),
                'merge_reasons': merge_reasons
            }
            merged.append(merged_det)
    
    return merged

def parse_xml_annotation(xml_path):
    """Parse XML annotation file to extract ground truth bounding boxes and labels"""
    try:
        tree = ET.parse(xml_path)
        root = tree.getroot()
        
        annotations = []
        image_info = {}
        
        # Get image dimensions
        size_elem = root.find('size')
        if size_elem is not None:
            try:
                image_info['width'] = int(size_elem.find('width').text)
                image_info['height'] = int(size_elem.find('height').text)
                image_info['depth'] = int(size_elem.find('depth').text)
            except:
                pass
        
        for obj in root.findall('object'):
            name_elem = obj.find('name')
            if name_elem is None or not name_elem.text:
                continue
                
            name = name_elem.text.strip()
            if not name:
                continue
                
            bbox = obj.find('bndbox')
            if bbox is None:
                continue
                
            try:
                x1 = int(bbox.find('xmin').text)
                y1 = int(bbox.find('ymin').text)
                x2 = int(bbox.find('xmax').text)
                y2 = int(bbox.find('ymax').text)
                
                annotations.append({
                    'label': name,
                    'bbox': [x1, y1, x2, y2],
                    'area': (x2 - x1) * (y2 - y1)
                })
            except (ValueError, AttributeError):
                continue
        
        return annotations, image_info
    except Exception as e:
        print(f"âŒ Error parsing {xml_path}: {e}")
        return [], {}

def analyze_dataset_statistics(image_files, images_folder, annotations_folder):
    """Analyze and display comprehensive dataset statistics"""
    print(f"\nğŸ“Š DATASET STATISTICS ANALYSIS")
    print("=" * 80)
    
    dataset_stats = {
        'images': [],
        'annotations': [],
        'labels': [],
        'bbox_areas': [],
        'bbox_aspect_ratios': [],
        'image_sizes': [],
        'objects_per_image': []
    }
    
    print(f"ğŸ” Analyzing {len(image_files)} images...")
    
    for image_file in tqdm(image_files, desc="Analyzing dataset"):
        image_path = os.path.join(images_folder, image_file)
        image_id = os.path.splitext(image_file)[0]
        xml_path = os.path.join(annotations_folder, f"{image_id}.xml")
        
        # Load image to get actual dimensions
        image = cv2.imread(image_path)
        if image is not None:
            h, w, c = image.shape
            dataset_stats['image_sizes'].append((w, h))
            dataset_stats['images'].append(image_file)
        
        # Parse annotations
        if os.path.exists(xml_path):
            annotations, image_info = parse_xml_annotation(xml_path)
            dataset_stats['objects_per_image'].append(len(annotations))
            
            for ann in annotations:
                dataset_stats['annotations'].append(ann)
                dataset_stats['labels'].append(ann['label'])
                
                # Calculate bbox statistics
                x1, y1, x2, y2 = ann['bbox']
                width = x2 - x1
                height = y2 - y1
                area = width * height
                aspect_ratio = width / height if height > 0 else 0
                
                dataset_stats['bbox_areas'].append(area)
                dataset_stats['bbox_aspect_ratios'].append(aspect_ratio)
    
    # Display comprehensive statistics
    print(f"\nğŸ“ˆ DATASET OVERVIEW:")
    print(f"   ğŸ“· Total Images: {len(dataset_stats['images'])}")
    print(f"   ğŸ“¦ Total Annotations: {len(dataset_stats['annotations'])}")
    print(f"   ğŸ·ï¸  Unique Labels: {len(set(dataset_stats['labels']))}")
    
    # Label distribution
    label_counts = Counter(dataset_stats['labels'])
    print(f"\nğŸ·ï¸  LABEL DISTRIBUTION:")
    for label, count in label_counts.most_common():
        percentage = (count / len(dataset_stats['labels'])) * 100
        print(f"   {label}: {count} ({percentage:.1f}%)")
    
    return dataset_stats

class GroundingDINODetector:
    """Grounding DINO detector for pedestrian behavior detection"""
    
    def __init__(self):
        # Setup device
        if torch.cuda.is_available():
            self.device = "cuda"
            print(f"âœ… GPU Available: {torch.cuda.get_device_name(0)}")
            torch.cuda.empty_cache()
        else:
            self.device = "cpu"
            print("âš ï¸ Using CPU")
        
        # Load Grounding DINO
        print("ğŸ”„ Loading Grounding DINO model...")
        model_id = "IDEA-Research/grounding-dino-tiny"
        self.processor = AutoProcessor.from_pretrained(model_id)
        self.model = AutoModelForZeroShotObjectDetection.from_pretrained(model_id)
        self.model = self.model.to(self.device)
        self.model.eval()
        print("âœ… Grounding DINO loaded successfully!")
    
    def detect_objects(self, image, classes=DETECTION_CLASSES, confidence_threshold=CONFIDENCE_THRESHOLD):
        """Detect multiple object classes in image using same logic as original pipeline"""
        # Convert to PIL if needed
        if isinstance(image, np.ndarray):
            if len(image.shape) == 3 and image.shape[2] == 3:
                pil_image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
            else:
                pil_image = Image.fromarray(image)
        else:
            pil_image = image
        
        width, height = pil_image.size
        total_area = width * height
        min_person_area = total_area * (MIN_PERSON_AREA_PERCENTAGE / 100)
        
        # Create text prompt for all classes
        text_prompt = [[class_name for class_name in classes]]
        
        # Prepare inputs
        inputs = self.processor(images=pil_image, text=text_prompt, return_tensors="pt")
        device_inputs = {k: v.to(self.device) if isinstance(v, torch.Tensor) else v 
                        for k, v in inputs.items()}
        
        # Run detection
        with torch.no_grad():
            outputs = self.model(**device_inputs)
        
        # Process results
        target_sizes = torch.tensor([[height, width]]).to(self.device)
        results = self.processor.post_process_grounded_object_detection(
            outputs,
            input_ids=device_inputs.get('input_ids'),
            box_threshold=confidence_threshold,
            text_threshold=0,
            target_sizes=target_sizes
        )[0]
        
        # Extract detections
        raw_detections = []
        if len(results['boxes']) > 0:
            boxes = results['boxes'].cpu().numpy()
            scores = results['scores'].cpu().numpy()
            
            for box, score in zip(boxes, scores):
                x1, y1, x2, y2 = box.astype(int)
                box_area = (x2 - x1) * (y2 - y1)
                
                detection = {
                    'box': [x1, y1, x2, y2],
                    'confidence': float(score),
                    'area': box_area,
                    'area_percentage': (box_area / total_area) * 100
                }
                raw_detections.append(detection)
        
        # Merge overlapping detections using modified logic (person with non-person only)
        merged_detections = merge_person_nonperson_boxes(
            raw_detections, IOU_THRESHOLD, CONTAINMENT_THRESHOLD
        )
        
        # Filter by area
        valid_detections = []
        for det in merged_detections:
            box_area = det['area'] if 'area' in det else (det['box'][2] - det['box'][0]) * (det['box'][3] - det['box'][1])
            area_percentage = (box_area / total_area) * 100
            
            # Update detection with area info
            det['area'] = box_area
            det['area_percentage'] = area_percentage
            
            # Only keep detections with sufficient area
            if area_percentage >= MIN_PERSON_AREA_PERCENTAGE:
                valid_detections.append(det)
        
        return valid_detections, pil_image









class QwenClassifier:
    """Memory-safe Qwen 2.5 VL classifier with aggressive image resizing and proper error handling"""
    
    def __init__(self, max_image_size=512):  # Reduced default size for better memory management
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        self.max_image_size = max_image_size  # Max dimension for input images
        print(f"ğŸ”„ Loading Qwen 2.5 VL model on {self.device}...")
        print(f"ğŸ“ Max image size set to: {max_image_size}px")
        
        # Clear GPU memory first
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        
        # Load Qwen 2.5 VL with memory optimizations
        model_id = "Qwen/Qwen2.5-VL-3B-Instruct"
        
        self.model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
            model_id, 
            torch_dtype=torch.float16,  # Use half precision to save memory
            device_map="auto"
        )
        
        self.processor = AutoProcessor.from_pretrained(model_id)
        
        # Import process_vision_info
        try:
            from qwen_vl_utils import process_vision_info
            self.process_vision_info = process_vision_info
        except ImportError:
            print("âš ï¸ Warning: qwen_vl_utils not available, using fallback")
            self.process_vision_info = None
        
        print("âœ… Qwen 2.5 VL loaded successfully!")
    
    def calculate_image_memory_usage(self, width, height):
        """Calculate approximate memory usage for an image"""
        # Rough estimation: 3 channels * 2 bytes (float16) * width * height
        # Plus overhead for processing
        base_memory = width * height * 3 * 2  # bytes
        processing_overhead = base_memory * 4  # 4x overhead for processing
        total_mb = (base_memory + processing_overhead) / (1024 * 1024)
        return total_mb
    
    def aggressive_resize_for_crops(self, pil_image):
        """Aggressively resize image crops to prevent OOM, especially for large crops"""
        width, height = pil_image.size
        pixels = width * height
        
        # Calculate memory usage
        estimated_memory_mb = self.calculate_image_memory_usage(width, height)
        
        print(f"   ğŸ“ Original crop: {width}Ã—{height} ({pixels:,} pixels, ~{estimated_memory_mb:.1f}MB)")
        
        # Define memory thresholds
        if estimated_memory_mb > 200:  # Very large crop
            target_size = 256
            print(f"   ğŸš¨ Very large crop detected! Resizing to {target_size}px max")
        elif estimated_memory_mb > 100:  # Large crop
            target_size = 384
            print(f"   âš ï¸ Large crop detected! Resizing to {target_size}px max")
        elif estimated_memory_mb > 50:  # Medium crop
            target_size = 512
            print(f"   ğŸ“ Medium crop detected! Resizing to {target_size}px max")
        else:
            target_size = self.max_image_size
        
        # Resize if needed
        max_dim = max(width, height)
        if max_dim > target_size:
            # Calculate new dimensions maintaining aspect ratio
            if width > height:
                new_width = target_size
                new_height = int(height * (target_size / width))
            else:
                new_height = target_size
                new_width = int(width * (target_size / height))
            
            print(f"   ğŸ“ Resizing: {width}Ã—{height} â†’ {new_width}Ã—{new_height}")
            pil_image = pil_image.resize((new_width, new_height), Image.Resampling.LANCZOS)
            
            # Verify new size
            new_memory = self.calculate_image_memory_usage(new_width, new_height)
            print(f"   âœ… New size: {new_width}Ã—{new_height} (~{new_memory:.1f}MB)")
        
        return pil_image
    
    def get_memory_info(self):
        """Get current GPU memory usage"""
        if torch.cuda.is_available():
            allocated = torch.cuda.memory_allocated() / 1024**3
            reserved = torch.cuda.memory_reserved() / 1024**3
            return f"GPU Memory: {allocated:.2f}GB allocated, {reserved:.2f}GB reserved"
        return "No GPU available"
    
    def classify_behavior(self, image):
        """Classify behavior using Qwen 2.5 VL with memory safety and proper error handling"""
        behavior_classes = [
            "walking", "running", "pushing a stroller", 
            "biking", "standing", "skateboarding", "scootering", 
            "person on wheelchair"
        ]
        
        prompt = f"""Describe the activity of the pedestrian based on posture, movement, and interaction with their environment or equipment. Focus on functional cues like body position, motion, use of hands or feet, and any relevant mobility devices (e.g., bike, scooter, stroller, wheelchair). Ignore unrelated visual details such as clothing style or colors unless they directly support the activity. Choose from these options:
{', '.join(behavior_classes)}

Important: Respond with ONLY ONE WORD from the list above. Do not add any other text.

Classification:"""
        
        # Initialize variables for proper cleanup
        inputs = None
        generated_ids = None
        
        try:
            # Convert to PIL if needed
            if isinstance(image, np.ndarray):
                if len(image.shape) == 3 and image.shape[2] == 3:
                    pil_image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
                else:
                    pil_image = Image.fromarray(image)
            else:
                pil_image = image
            
            # Aggressive resize for crops (especially large ones)
            pil_image = self.aggressive_resize_for_crops(pil_image)
            
            # Clear GPU cache before processing
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
            
            # Log memory before processing
            print(f"   ğŸ’¾ Before processing: {self.get_memory_info()}")
            
            # Prepare messages for Qwen
            messages = [
                {
                    "role": "user",
                    "content": [
                        {
                            "type": "image",
                            "image": pil_image,
                        },
                        {"type": "text", "text": prompt},
                    ],
                }
            ]
            
            # Apply chat template
            text = self.processor.apply_chat_template(
                messages, tokenize=False, add_generation_prompt=True
            )
            
            # Process vision info
            if self.process_vision_info:
                image_inputs, video_inputs = self.process_vision_info(messages)
            else:
                image_inputs = [pil_image]
                video_inputs = None
            
            # Process inputs
            inputs = self.processor(
                text=[text],
                images=image_inputs,
                videos=video_inputs,
                padding=True,
                return_tensors="pt",
            )
            inputs = inputs.to(self.device)
            
            # Generate response with memory optimization
            with torch.no_grad():
                generated_ids = self.model.generate(
                    **inputs,
                    max_new_tokens=10,
                    do_sample=False  # Deterministic generation
                )
            
            # Decode the response
            generated_ids_trimmed = [
                out_ids[len(in_ids):] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
            ]
            
            response = self.processor.batch_decode(
                generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
            )[0]
            
            # Clean up response
            classification = response.strip().lower()
            
            # Log memory after processing
            print(f"   ğŸ’¾ After processing: {self.get_memory_info()}")
            
            # Map to predefined classes
            for behavior in behavior_classes:
                if behavior.lower() in classification:
                    return behavior, 1.0, {"labels": [behavior], "scores": [1.0]}
            
            # If no exact match, return the raw classification
            return classification, 1.0, {"labels": [classification], "scores": [1.0]}
            
        except torch.cuda.OutOfMemoryError as e:
            print(f"ğŸ’¥ CUDA OOM Error: {e}")
            print(f"   ğŸ’¾ {self.get_memory_info()}")
            
            # Emergency memory cleanup
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
                torch.cuda.ipc_collect()
            
            # Try with emergency tiny image
            try:
                print(f"   ğŸ”„ Emergency retry with 128px max...")
                if isinstance(image, np.ndarray):
                    emergency_image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
                else:
                    emergency_image = image
                
                # Force very small size
                emergency_image.thumbnail((128, 128), Image.Resampling.LANCZOS)
                print(f"   ğŸ“ Emergency size: {emergency_image.size}")
                
                # Try classification again with tiny image
                return self.classify_with_tiny_image(emergency_image, prompt, behavior_classes)
                
            except Exception as retry_error:
                print(f"   âŒ Emergency retry failed: {retry_error}")
                return "walking", 0.1, {"labels": ["walking"], "scores": [0.1]}  # Safe fallback
            
        except Exception as e:
            print(f"âŒ Error classifying behavior: {e}")
            return "unknown", 0.0, {"labels": ["unknown"], "scores": [0.0]}
        
        finally:
            # Clean up safely - only delete if variables exist
            try:
                if inputs is not None:
                    del inputs
                if generated_ids is not None:
                    del generated_ids
                if torch.cuda.is_available():
                    torch.cuda.empty_cache()
            except Exception as cleanup_error:
                print(f"   âš ï¸ Cleanup warning: {cleanup_error}")
    
    def classify_with_tiny_image(self, tiny_image, prompt, behavior_classes):
        """Emergency classification with very small image"""
        try:
            # Prepare minimal processing
            messages = [
                {
                    "role": "user",
                    "content": [
                        {"type": "image", "image": tiny_image},
                        {"type": "text", "text": prompt},
                    ],
                }
            ]
            
            text = self.processor.apply_chat_template(
                messages, tokenize=False, add_generation_prompt=True
            )
            
            if self.process_vision_info:
                image_inputs, video_inputs = self.process_vision_info(messages)
            else:
                image_inputs = [tiny_image]
                video_inputs = None
            
            inputs = self.processor(
                text=[text],
                images=image_inputs,
                videos=video_inputs,
                padding=True,
                return_tensors="pt",
            )
            inputs = inputs.to(self.device)
            
            with torch.no_grad():
                generated_ids = self.model.generate(
                    **inputs,
                    max_new_tokens=5,  # Even fewer tokens
                    do_sample=False
                )
            
            generated_ids_trimmed = [
                out_ids[len(in_ids):] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
            ]
            
            response = self.processor.batch_decode(
                generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
            )[0]
            
            classification = response.strip().lower()
            
            # Map to predefined classes
            for behavior in behavior_classes:
                if behavior.lower() in classification:
                    return behavior, 0.5, {"labels": [behavior], "scores": [0.5]}  # Lower confidence
            
            return "walking", 0.2, {"labels": ["walking"], "scores": [0.2]}  # Default fallback
            
        except Exception as e:
            print(f"   âŒ Tiny image classification failed: {e}")
            return "walking", 0.1, {"labels": ["walking"], "scores": [0.1]}
        
        finally:
            if torch.cuda.is_available():
                torch.cuda.empty_cache()

# Set environment variable to help with memory fragmentation
import os
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

print("ğŸš€ FIXED MEMORY-SAFE QWEN CLASSIFIER READY!")
print("âœ… Aggressive crop resizing based on memory usage")
print("âœ… Proper error handling and variable cleanup")
print("âœ… Emergency fallback for OOM situations")
print("âœ… Memory usage estimation and monitoring")
print("ğŸ“ Default max image size: 512px (conservative)")
print("ğŸ”§ Set PYTORCH_CUDA_ALLOC_CONF=expandable_segments:True")






def map_detections_to_annotations(detections, annotations, iou_threshold=IOU_MAPPING_THRESHOLD):
    """Map detected bboxes to ground truth annotations using IoU"""
    mapped_pairs = []
    used_detections = set()
    used_annotations = set()
    
    for det_idx, detection in enumerate(detections):
        best_iou = 0
        best_ann_idx = -1
        
        for ann_idx, annotation in enumerate(annotations):
            if ann_idx in used_annotations:
                continue
                
            iou = calculate_iou(detection['box'], annotation['bbox'])
            if iou >= iou_threshold and iou > best_iou:
                best_iou = iou
                best_ann_idx = ann_idx
        
        if best_ann_idx != -1:
            mapped_pairs.append({
                'detection_idx': det_idx,
                'annotation_idx': best_ann_idx,
                'detection': detection,
                'annotation': annotations[best_ann_idx],
                'iou': best_iou
            })
            used_detections.add(det_idx)
            used_annotations.add(best_ann_idx)
    
    # Find unmatched detections and annotations
    unmatched_detections = [i for i in range(len(detections)) if i not in used_detections]
    unmatched_annotations = [i for i in range(len(annotations)) if i not in used_annotations]
    
    return mapped_pairs, unmatched_detections, unmatched_annotations

def compute_detection_metrics(mapped_pairs, total_detections, total_annotations):
    """Compute detection performance metrics"""
    true_positives = len(mapped_pairs)
    false_positives = total_detections - true_positives
    false_negatives = total_annotations - true_positives
    
    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    return {
        'true_positives': true_positives,
        'false_positives': false_positives,
        'false_negatives': false_negatives,
        'precision': precision,
        'recall': recall,
        'f1_score': f1_score
    }

def display_comprehensive_metrics(all_results):
    """Display comprehensive classification metrics with enhanced analysis"""
    if not all_results:
        print("âŒ No results to display")
        return None, None
    
    df = pd.DataFrame(all_results)
    
    # Overall accuracy
    overall_accuracy = accuracy_score(df['true_label'], df['predicted_label'])
    print(f"\nğŸ¯ OVERALL CLASSIFICATION ACCURACY: {overall_accuracy:.3f} ({overall_accuracy*100:.1f}%)")
    
    # Get unique labels
    all_labels = sorted(set(df['true_label'].tolist() + df['predicted_label'].tolist()))
    
    # Classification report
    print(f"\nğŸ“Š DETAILED CLASSIFICATION REPORT:")
    print("="*80)
    report = classification_report(df['true_label'], df['predicted_label'], 
                                 output_dict=False, zero_division=0)
    print(report)
    
    # Get classification report as dict for further analysis
    report_dict = classification_report(df['true_label'], df['predicted_label'], 
                                      output_dict=True, zero_division=0)
    
    # Per-class metrics with confidence intervals
    precision, recall, f1, support = precision_recall_fscore_support(
        df['true_label'], df['predicted_label'], average=None, zero_division=0, labels=all_labels
    )
    
    metrics_df = pd.DataFrame({
        'Class': all_labels,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Support': support
    })
    
    print(f"\nğŸ“ˆ ENHANCED PER-CLASS METRICS:")
    print("="*80)
    print(metrics_df.round(3).to_string(index=False))
    
    # Class-wise confusion matrices
    cm = confusion_matrix(df['true_label'], df['predicted_label'], labels=all_labels)
    
    # Create comprehensive visualizations
    fig = plt.figure(figsize=(20, 12))
    
    # 1. Overall Confusion Matrix
    plt.subplot(2, 3, 1)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=all_labels, yticklabels=all_labels)
    plt.title('Overall Confusion Matrix', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    
    # 2. Normalized Confusion Matrix
    plt.subplot(2, 3, 2)
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=all_labels, yticklabels=all_labels)
    plt.title('Normalized Confusion Matrix', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    
    # 3. Per-class F1 scores
    plt.subplot(2, 3, 3)
    plt.bar(all_labels, f1, color='lightcoral', alpha=0.7)
    plt.title('F1-Score by Class')
    plt.xlabel('Class')
    plt.ylabel('F1-Score')
    plt.xticks(rotation=45)
    plt.ylim(0, 1)
    for i, v in enumerate(f1):
        plt.text(i, v + 0.01, f'{v:.3f}', ha='center', va='bottom')
    
    # 4. IoU distribution
    plt.subplot(2, 3, 4)
    iou_values = df['iou']
    plt.hist(iou_values, bins=20, alpha=0.7, color='blue')
    plt.xlabel('IoU Score')
    plt.ylabel('Frequency')
    plt.title('IoU Distribution')
    plt.axvline(iou_values.mean(), color='red', linestyle='--', 
                label=f'Mean: {iou_values.mean():.3f}')
    plt.legend()
    
    # 5. Class distribution (true vs predicted)
    plt.subplot(2, 3, 5)
    true_counts = df['true_label'].value_counts()
    pred_counts = df['predicted_label'].value_counts()
    
    x = np.arange(len(all_labels))
    width = 0.35
    
    true_vals = [true_counts.get(label, 0) for label in all_labels]
    pred_vals = [pred_counts.get(label, 0) for label in all_labels]
    
    plt.bar(x - width/2, true_vals, width, label='True', alpha=0.7, color='skyblue')
    plt.bar(x + width/2, pred_vals, width, label='Predicted', alpha=0.7, color='orange')
    plt.xlabel('Class')
    plt.ylabel('Count')
    plt.title('Class Distribution: True vs Predicted')
    plt.xticks(x, all_labels, rotation=45)
    plt.legend()
    
    # 6. Area analysis
    plt.subplot(2, 3, 6)
    area_percentages = df['area_percentage']
    plt.hist(area_percentages, bins=20, alpha=0.7, color='purple')
    plt.xlabel('Area Percentage of Image')
    plt.ylabel('Frequency')
    plt.title('Detection Area Distribution')
    plt.axvline(area_percentages.mean(), color='red', linestyle='--', 
                label=f'Mean: {area_percentages.mean():.1f}%')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Print detailed error analysis
    print(f"\nğŸ” DETAILED ERROR ANALYSIS:")
    print("="*80)
    
    # Most confused pairs
    error_pairs = []
    for i, true_label in enumerate(all_labels):
        for j, pred_label in enumerate(all_labels):
            if i != j and cm[i, j] > 0:
                error_pairs.append((true_label, pred_label, cm[i, j]))
    
    error_pairs.sort(key=lambda x: x[2], reverse=True)
    
    print("Top 5 Most Confused Class Pairs:")
    for i, (true_label, pred_label, count) in enumerate(error_pairs[:5]):
        print(f"   {i+1}. {true_label} â†’ {pred_label}: {count} errors")
    
    # Best and worst performing classes
    best_class_idx = np.argmax(f1)
    worst_class_idx = np.argmin(f1)
    
    print(f"\nBest Performing Class: {all_labels[best_class_idx]} (F1: {f1[best_class_idx]:.3f})")
    print(f"Worst Performing Class: {all_labels[worst_class_idx]} (F1: {f1[worst_class_idx]:.3f})")
    
    # IoU statistics
    print(f"\nIoU Statistics:")
    print(f"   Mean IoU: {df['iou'].mean():.3f}")
    print(f"   IoU Std: {df['iou'].std():.3f}")
    print(f"   Min IoU: {df['iou'].min():.3f}")
    print(f"   Max IoU: {df['iou'].max():.3f}")
    
    return df, metrics_df

def run_pedestrian_behavior_pipeline():
    """Main pipeline for pedestrian behavior dynamics analysis with Qwen 2.5 VL"""
    print("ğŸš€ PEDESTRIAN BEHAVIOR DYNAMICS PIPELINE - GROUNDING DINO + QWEN 2.5 VL")
    print("=" * 80)
    print(f"ğŸ¯ Detection Classes: {', '.join(DETECTION_CLASSES)}")
    print(f"ğŸ¯ Confidence Threshold: {CONFIDENCE_THRESHOLD}")
    print(f"ğŸ”„ IoU Merge Threshold: {IOU_THRESHOLD}")
    print(f"ğŸ“ Containment Threshold: {CONTAINMENT_THRESHOLD}")
    print(f"ğŸ“Š Min Area Percentage: {MIN_PERSON_AREA_PERCENTAGE}%")
    print(f"ğŸ”— IoU Mapping Threshold: {IOU_MAPPING_THRESHOLD}")
    print("=" * 80)
    
    # Setup environment
    setup_environment()
    
    # Create output directory
    os.makedirs(OUTPUT_FOLDER, exist_ok=True)
    
    # Find image files
    image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
    image_files = []
    
    # Check if folders exist
    if not os.path.exists(IMAGES_FOLDER):
        print(f"âŒ Images folder not found: {IMAGES_FOLDER}")
        return None
    
    if not os.path.exists(ANNOTATIONS_FOLDER):
        print(f"âŒ Annotations folder not found: {ANNOTATIONS_FOLDER}")
        return None
    
    for file in os.listdir(IMAGES_FOLDER):
        if any(file.lower().endswith(ext) for ext in image_extensions):
            image_files.append(file)
    
    if not image_files:
        print(f"âŒ No image files found in {IMAGES_FOLDER}")
        return None
    
    print(f"ğŸ–¼ï¸ Found {len(image_files)} images")
    
    # Analyze dataset statistics
    dataset_stats = analyze_dataset_statistics(image_files, IMAGES_FOLDER, ANNOTATIONS_FOLDER)
    
    # Initialize models
    detector = GroundingDINODetector()
    classifier = QwenClassifier()
    
    # Pipeline results storage
    all_results = []
    detection_metrics = []
    all_classifications = []
    
    print(f"\nğŸ”„ Processing {len(image_files)} images...")
    
    for i, image_file in enumerate(tqdm(image_files, desc="Processing images")):
        image_path = os.path.join(IMAGES_FOLDER, image_file)
        image_id = os.path.splitext(image_file)[0]
        xml_path = os.path.join(ANNOTATIONS_FOLDER, f"{image_id}.xml")
        
        print(f"\n{'='*60}")
        print(f"ğŸ“· IMAGE {i+1}/{len(image_files)}: {image_file}")
        
        if not os.path.exists(xml_path):
            print(f"âŒ No annotation found for {image_file}")
            continue
        
        # Load image and annotation
        image = cv2.imread(image_path)
        if image is None:
            print(f"âŒ Could not load image: {image_file}")
            continue
        
        annotations, image_info = parse_xml_annotation(xml_path)
        if not annotations:
            print(f"âŒ No valid annotations found in {xml_path}")
            continue
        
        print(f"ğŸ“‹ Ground truth annotations: {len(annotations)}")
        for j, ann in enumerate(annotations):
            print(f"   GT {j+1}: {ann['label']} at {ann['bbox']}")
        
        # Run Grounding DINO detection
        detections, pil_image = detector.detect_objects(image)
        print(f"ğŸ” Grounding DINO detections: {len(detections)}")
        
        # Map detections to annotations
        mapped_pairs, unmatched_detections, unmatched_annotations = map_detections_to_annotations(
            detections, annotations, IOU_MAPPING_THRESHOLD
        )
        
        print(f"ğŸ”— Mapping results:")
        print(f"   âœ… Mapped pairs: {len(mapped_pairs)}")
        print(f"   ğŸ“¦ Unmatched detections: {len(unmatched_detections)} (discarded)")
        print(f"   ğŸ“‹ Unmatched annotations: {len(unmatched_annotations)}")
        
        # Compute detection metrics
        det_metrics = compute_detection_metrics(mapped_pairs, len(detections), len(annotations))
        det_metrics['image_file'] = image_file
        detection_metrics.append(det_metrics)
        
        print(f"ğŸ“Š Detection Metrics - P: {det_metrics['precision']:.3f}, R: {det_metrics['recall']:.3f}, F1: {det_metrics['f1_score']:.3f}")
        
        # Process only mapped pairs for classification
        print(f"ğŸ§  Running Qwen 2.5 VL classification on {len(mapped_pairs)} mapped detections...")
        
        for pair_idx, pair in enumerate(mapped_pairs):
            detection = pair['detection']
            annotation = pair['annotation']
            x1, y1, x2, y2 = detection['box']
            
            # Crop detection
            img_array = np.array(pil_image)
            crop = img_array[y1:y2, x1:x2]
            
            if crop.size > 0:
                # Classify behavior with Qwen
                pred_label, pred_conf, all_scores = classifier.classify_behavior(crop)
                all_classifications.append(pred_label)
                
                # Store comprehensive results
                result = {
                    'image_file': image_file,
                    'detection_idx': pair['detection_idx'],
                    'annotation_idx': pair['annotation_idx'],
                    'true_label': annotation['label'],
                    'predicted_label': pred_label,
                    'confidence': pred_conf,
                    'qwen_classification': pred_label,
                    'iou': pair['iou'],
                    'detection_bbox': detection['box'],
                    'annotation_bbox': annotation['bbox'],
                    'detection_confidence': detection['confidence'],
                    'detection_area': detection['area'],
                    'area_percentage': detection['area_percentage'],
                    'correct': pred_label == annotation['label']
                }
                all_results.append(result)
                
                # Print classification result
                print(f"   ğŸ¯ Crop {pair_idx+1}:")
                print(f"      ğŸ·ï¸  Classification: {annotation['label']} -> {pred_label} | IoU: {pair['iou']:.3f}")
        
        # Clean GPU memory
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
    
    # Save detailed results
    if all_results:
        results_df = pd.DataFrame(all_results)
        results_df.to_csv(os.path.join(OUTPUT_FOLDER, 'pedestrian_behavior_qwen_results.csv'), index=False)
        
        detection_df = pd.DataFrame(detection_metrics)
        detection_df.to_csv(os.path.join(OUTPUT_FOLDER, 'detection_metrics.csv'), index=False)
        
        # Save classifications
        classifications_df = pd.DataFrame({'classifications': all_classifications})
        classifications_df.to_csv(os.path.join(OUTPUT_FOLDER, 'qwen_classifications.csv'), index=False)
        
        # Display comprehensive metrics
        print(f"\nğŸ‰ PIPELINE COMPLETE!")
        print(f"ğŸ’¾ Results saved to: {OUTPUT_FOLDER}")
        print(f"ğŸ“Š Total processed detections: {len(all_results)}")
        print(f"ğŸ“ Total Qwen classifications: {len(all_classifications)}")
        
        # Detection Performance Summary
        if detection_metrics:
            avg_precision = np.mean([m['precision'] for m in detection_metrics])
            avg_recall = np.mean([m['recall'] for m in detection_metrics])
            avg_f1 = np.mean([m['f1_score'] for m in detection_metrics])
            
            print(f"\nğŸ” DETECTION PERFORMANCE SUMMARY:")
            print(f"   ğŸ“ˆ Average Precision: {avg_precision:.3f}")
            print(f"   ğŸ“ˆ Average Recall: {avg_recall:.3f}")
            print(f"   ğŸ“ˆ Average F1-Score: {avg_f1:.3f}")
        
        # Classification Performance with enhanced metrics
        final_df, metrics_df = display_comprehensive_metrics(all_results)
        
        # Display sample classifications
        print(f"\nğŸ“ SAMPLE QWEN CLASSIFICATIONS:")
        print("="*60)
        for i, classification in enumerate(all_classifications[:10]):
            print(f"   {i+1}. {classification}")
        if len(all_classifications) > 10:
            print(f"   ... and {len(all_classifications)-10} more classifications")
        
        return final_df, detection_df, metrics_df, dataset_stats
    
    else:
        print("âŒ No results to display")
        return None, None, None, None

# Run the complete pipeline
print("ğŸš€ Starting Pedestrian Behavior Dynamics Pipeline with Qwen 2.5 VL...")
results_df, detection_df, class_metrics_df, dataset_stats = run_pedestrian_behavior_pipeline()
```
