```python
import os
import sys
import subprocess
import cv2
import torch
import numpy as np
import pandas as pd
from tqdm import tqdm
from PIL import Image, ImageEnhance
from transformers import (
    AutoProcessor, 
    AutoModelForZeroShotObjectDetection,
    BlipProcessor,
    BlipForConditionalGeneration,
    pipeline
)
import json
import xml.etree.ElementTree as ET
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from sklearn.metrics import (
    classification_report, 
    confusion_matrix, 
    accuracy_score,
    precision_recall_fscore_support,
    precision_recall_curve,
    roc_curve,
    auc
)
import seaborn as sns
from collections import defaultdict, Counter
import warnings
warnings.filterwarnings('ignore')


DATASET_FOLDER = '/kaggle/input/pedestrian-behavior-on-sidewalk'
IMAGES_FOLDER = '/kaggle/input/pedestrian-behavior-on-sidewalk' 
ANNOTATIONS_FOLDER = '/kaggle/input/pedestrian-behavior-on-sidewalk'  
OUTPUT_FOLDER = '/kaggle/working/pipeline_results'
CONFIDENCE_THRESHOLD = 0.5
IOU_THRESHOLD = 0.1
CONTAINMENT_THRESHOLD = 0.3
MIN_PERSON_AREA_PERCENTAGE = 0.5
IOU_MAPPING_THRESHOLD = 0.8

# Detection classes
DETECTION_CLASSES = ["person", "bike", "skateboard", "baby stroller", "wheelchair"]

def setup_environment():
    """Install required packages"""
    print("üöÄ Setting up Pedestrian Behavior Dynamics Pipeline...")
    packages = [
        "torch torchvision transformers", 
        "opencv-python-headless", 
        "tqdm", 
        "matplotlib", 
        "seaborn",
        "scikit-learn",
        "pandas"
    ]
    
    for package in packages:
        print(f"üì¶ Installing {package}...")
        result = subprocess.run(f"pip install {package}", shell=True, capture_output=True, text=True)
        if result.returncode == 0:
            print(f"‚úÖ {package} installed successfully")

def calculate_iou(box1, box2):
    """Calculate Intersection over Union (IoU) between two bounding boxes"""
    x1_1, y1_1, x2_1, y2_1 = box1
    x1_2, y1_2, x2_2, y2_2 = box2
    
    # Calculate intersection
    x1_i = max(x1_1, x1_2)
    y1_i = max(y1_1, y1_2)
    x2_i = min(x2_1, x2_2)
    y2_i = min(y2_1, y2_2)
    
    if x2_i <= x1_i or y2_i <= y1_i:
        return 0.0
    
    intersection = (x2_i - x1_i) * (y2_i - y1_i)
    area1 = (x2_1 - x1_1) * (y2_1 - y1_1)
    area2 = (x2_2 - x1_2) * (y2_2 - y1_2)
    union = area1 + area2 - intersection
    
    return intersection / union if union > 0 else 0.0

def calculate_containment(box1, box2):
    """Calculate what percentage of box1 is contained within box2"""
    x1_1, y1_1, x2_1, y2_1 = box1
    x1_2, y1_2, x2_2, y2_2 = box2
    
    # Calculate intersection
    x1_i = max(x1_1, x1_2)
    y1_i = max(y1_1, y1_2)
    x2_i = min(x2_1, x2_2)
    y2_i = min(y2_1, y2_2)
    
    if x2_i <= x1_i or y2_i <= y1_i:
        return 0.0
    
    intersection = (x2_i - x1_i) * (y2_i - y1_i)
    area1 = (x2_1 - x1_1) * (y2_1 - y1_1)
    
    return intersection / area1 if area1 > 0 else 0.0

def should_merge_person_nonperson(det1, det2, iou_threshold=0.1, containment_threshold=0.7):
    """
    Modified merging logic: Only merge person with non-person labels
    Do NOT merge two person bounding boxes
    """
    box1, box2 = det1['box'], det2['box']
    
    # Calculate IoU
    iou = calculate_iou(box1, box2)
    if iou > iou_threshold:
        return True, "iou"
    
    # Calculate areas
    area1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area2 = (box2[2] - box2[0]) * (box2[3] - box2[1])
    
    # Check containment for smaller objects (like skateboards in person boxes)
    if area1 < area2:  # box1 is smaller
        containment = calculate_containment(box1, box2)
        if containment > containment_threshold:
            return True, f"containment_{containment:.2f}"
    else:  # box2 is smaller
        containment = calculate_containment(box2, box1)
        if containment > containment_threshold:
            return True, f"containment_{containment:.2f}"
    
    return False, "no_merge"

def merge_person_nonperson_boxes(detections, iou_threshold=0.1, containment_threshold=0.7):
    """
    Merge overlapping bounding boxes but ONLY merge person with non-person
    Keep all person boxes separate from each other
    """
    if len(detections) <= 1:
        return detections
    
    merged = []
    used = set()
    
    for i, det1 in enumerate(detections):
        if i in used:
            continue
            
        overlapping = [det1]
        used.add(i)
        merge_reasons = []
        
        for j, det2 in enumerate(detections[i+1:], i+1):
            if j in used:
                continue
                
            should_merge, reason = should_merge_person_nonperson(det1, det2, iou_threshold, containment_threshold)
            
            if should_merge:
                overlapping.append(det2)
                merge_reasons.append(f"box_{j}_{reason}")
                used.add(j)
        
        if len(overlapping) == 1:
            merged.append(overlapping[0])
        else:
            # Merge overlapping detections - take the one with highest confidence
            best_det = max(overlapping, key=lambda x: x['confidence'])
            
            # Expand bounding box to encompass all overlapping boxes
            all_boxes = [det['box'] for det in overlapping]
            x1_min = min(box[0] for box in all_boxes)
            y1_min = min(box[1] for box in all_boxes)
            x2_max = max(box[2] for box in all_boxes)
            y2_max = max(box[3] for box in all_boxes)
            
            merged_det = {
                'box': [x1_min, y1_min, x2_max, y2_max],
                'confidence': best_det['confidence'],
                'merged_count': len(overlapping),
                'merge_reasons': merge_reasons
            }
            merged.append(merged_det)
    
    return merged

def parse_xml_annotation(xml_path):
    """Parse XML annotation file to extract ground truth bounding boxes and labels"""
    try:
        tree = ET.parse(xml_path)
        root = tree.getroot()
        
        annotations = []
        image_info = {}
        
        # Get image dimensions
        size_elem = root.find('size')
        if size_elem is not None:
            try:
                image_info['width'] = int(size_elem.find('width').text)
                image_info['height'] = int(size_elem.find('height').text)
                image_info['depth'] = int(size_elem.find('depth').text)
            except:
                pass
        
        for obj in root.findall('object'):
            name_elem = obj.find('name')
            if name_elem is None or not name_elem.text:
                continue  # Skip objects without names
                
            name = name_elem.text.strip()
            if not name:  # Skip empty names
                continue
                
            bbox = obj.find('bndbox')
            if bbox is None:
                continue
                
            try:
                x1 = int(bbox.find('xmin').text)
                y1 = int(bbox.find('ymin').text)
                x2 = int(bbox.find('xmax').text)
                y2 = int(bbox.find('ymax').text)
                
                annotations.append({
                    'label': name,
                    'bbox': [x1, y1, x2, y2],
                    'area': (x2 - x1) * (y2 - y1)
                })
            except (ValueError, AttributeError):
                continue
        
        return annotations, image_info
    except Exception as e:
        print(f"‚ùå Error parsing {xml_path}: {e}")
        return [], {}

def analyze_dataset_statistics(image_files, images_folder, annotations_folder):
    """Analyze and display comprehensive dataset statistics"""
    print(f"\nüìä DATASET STATISTICS ANALYSIS")
    print("=" * 80)
    
    dataset_stats = {
        'images': [],
        'annotations': [],
        'labels': [],
        'bbox_areas': [],
        'bbox_aspect_ratios': [],
        'image_sizes': [],
        'objects_per_image': []
    }
    
    print(f"üîç Analyzing {len(image_files)} images...")
    
    for image_file in tqdm(image_files, desc="Analyzing dataset"):
        image_path = os.path.join(images_folder, image_file)
        image_id = os.path.splitext(image_file)[0]
        xml_path = os.path.join(annotations_folder, f"{image_id}.xml")
        
        # Load image to get actual dimensions
        image = cv2.imread(image_path)
        if image is not None:
            h, w, c = image.shape
            dataset_stats['image_sizes'].append((w, h))
            dataset_stats['images'].append(image_file)
        
        # Parse annotations
        if os.path.exists(xml_path):
            annotations, image_info = parse_xml_annotation(xml_path)
            dataset_stats['objects_per_image'].append(len(annotations))
            
            for ann in annotations:
                dataset_stats['annotations'].append(ann)
                dataset_stats['labels'].append(ann['label'])
                
                # Calculate bbox statistics
                x1, y1, x2, y2 = ann['bbox']
                width = x2 - x1
                height = y2 - y1
                area = width * height
                aspect_ratio = width / height if height > 0 else 0
                
                dataset_stats['bbox_areas'].append(area)
                dataset_stats['bbox_aspect_ratios'].append(aspect_ratio)
    
    # Display comprehensive statistics
    print(f"\nüìà DATASET OVERVIEW:")
    print(f"   üì∑ Total Images: {len(dataset_stats['images'])}")
    print(f"   üì¶ Total Annotations: {len(dataset_stats['annotations'])}")
    print(f"   üè∑Ô∏è  Unique Labels: {len(set(dataset_stats['labels']))}")
    
    # Label distribution
    label_counts = Counter(dataset_stats['labels'])
    print(f"\nüè∑Ô∏è  LABEL DISTRIBUTION:")
    for label, count in label_counts.most_common():
        percentage = (count / len(dataset_stats['labels'])) * 100
        print(f"   {label}: {count} ({percentage:.1f}%)")
    
    # Image statistics
    if dataset_stats['image_sizes']:
        widths = [size[0] for size in dataset_stats['image_sizes']]
        heights = [size[1] for size in dataset_stats['image_sizes']]
        
        print(f"\nüìê IMAGE DIMENSIONS:")
        print(f"   Width  - Mean: {np.mean(widths):.0f}, Std: {np.std(widths):.0f}, Range: {min(widths)}-{max(widths)}")
        print(f"   Height - Mean: {np.mean(heights):.0f}, Std: {np.std(heights):.0f}, Range: {min(heights)}-{max(heights)}")
    
    # Objects per image
    if dataset_stats['objects_per_image']:
        obj_per_img = dataset_stats['objects_per_image']
        print(f"\nüìä OBJECTS PER IMAGE:")
        print(f"   Mean: {np.mean(obj_per_img):.1f}, Std: {np.std(obj_per_img):.1f}")
        print(f"   Min: {min(obj_per_img)}, Max: {max(obj_per_img)}, Median: {np.median(obj_per_img):.1f}")
    
    # Bounding box statistics
    if dataset_stats['bbox_areas']:
        areas = dataset_stats['bbox_areas']
        aspects = dataset_stats['bbox_aspect_ratios']
        
        print(f"\nüìè BOUNDING BOX STATISTICS:")
        print(f"   Area - Mean: {np.mean(areas):.0f}, Std: {np.std(areas):.0f}")
        print(f"   Area - Min: {min(areas):.0f}, Max: {max(areas):.0f}, Median: {np.median(areas):.0f}")
        print(f"   Aspect Ratio - Mean: {np.mean(aspects):.2f}, Std: {np.std(aspects):.2f}")
        print(f"   Aspect Ratio - Min: {min(aspects):.2f}, Max: {max(aspects):.2f}")
    
    # Create visualizations
    plt.figure(figsize=(20, 12))
    
    # Label distribution
    plt.subplot(2, 4, 1)
    labels, counts = zip(*label_counts.most_common())
    plt.bar(labels, counts, color='skyblue')
    plt.title('Label Distribution')
    plt.xlabel('Labels')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    
    # Objects per image distribution
    plt.subplot(2, 4, 2)
    plt.hist(dataset_stats['objects_per_image'], bins=20, color='lightgreen', alpha=0.7)
    plt.title('Objects per Image Distribution')
    plt.xlabel('Number of Objects')
    plt.ylabel('Frequency')
    
    # Bounding box area distribution
    plt.subplot(2, 4, 3)
    plt.hist(dataset_stats['bbox_areas'], bins=30, color='coral', alpha=0.7)
    plt.title('Bounding Box Area Distribution')
    plt.xlabel('Area (pixels¬≤)')
    plt.ylabel('Frequency')
    plt.yscale('log')
    
    # Aspect ratio distribution
    plt.subplot(2, 4, 4)
    plt.hist(dataset_stats['bbox_aspect_ratios'], bins=30, color='gold', alpha=0.7)
    plt.title('Aspect Ratio Distribution')
    plt.xlabel('Width/Height Ratio')
    plt.ylabel('Frequency')
    
    # Image size scatter plot
    plt.subplot(2, 4, 5)
    if dataset_stats['image_sizes']:
        widths, heights = zip(*dataset_stats['image_sizes'])
        plt.scatter(widths, heights, alpha=0.6, color='purple')
        plt.title('Image Dimensions')
        plt.xlabel('Width')
        plt.ylabel('Height')
    
    # Box plot for areas by label
    plt.subplot(2, 4, 6)
    if len(set(dataset_stats['labels'])) > 1:
        label_areas = defaultdict(list)
        for ann in dataset_stats['annotations']:
            label_areas[ann['label']].append(ann['area'])
        
        labels_for_plot = []
        areas_for_plot = []
        for label, areas in label_areas.items():
            labels_for_plot.extend([label] * len(areas))
            areas_for_plot.extend(areas)
        
        df_box = pd.DataFrame({'Label': labels_for_plot, 'Area': areas_for_plot})
        sns.boxplot(data=df_box, x='Label', y='Area')
        plt.title('Bounding Box Areas by Label')
        plt.xticks(rotation=45)
        plt.yscale('log')
    
    # Cumulative distribution of areas
    plt.subplot(2, 4, 7)
    sorted_areas = np.sort(dataset_stats['bbox_areas'])
    cumulative = np.arange(1, len(sorted_areas) + 1) / len(sorted_areas)
    plt.plot(sorted_areas, cumulative, color='red', linewidth=2)
    plt.title('Cumulative Area Distribution')
    plt.xlabel('Area (pixels¬≤)')
    plt.ylabel('Cumulative Probability')
    plt.xscale('log')
    
    # Label co-occurrence matrix (if multiple objects per image)
    plt.subplot(2, 4, 8)
    if len(set(dataset_stats['labels'])) > 1:
        # Create co-occurrence matrix
        unique_labels = sorted(set(dataset_stats['labels']))
        cooccur_matrix = np.zeros((len(unique_labels), len(unique_labels)))
        
        # Group annotations by image
        image_annotations = defaultdict(list)
        for i, image_file in enumerate(dataset_stats['images']):
            image_id = os.path.splitext(image_file)[0]
            xml_path = os.path.join(annotations_folder, f"{image_id}.xml")
            if os.path.exists(xml_path):
                annotations, _ = parse_xml_annotation(xml_path)
                image_labels = [ann['label'] for ann in annotations]
                for label1 in image_labels:
                    for label2 in image_labels:
                        i1 = unique_labels.index(label1)
                        i2 = unique_labels.index(label2)
                        cooccur_matrix[i1, i2] += 1
        
        sns.heatmap(cooccur_matrix, annot=True, fmt='.0f', 
                   xticklabels=unique_labels, yticklabels=unique_labels,
                   cmap='Blues')
        plt.title('Label Co-occurrence Matrix')
        plt.xlabel('Co-occurring Labels')
        plt.ylabel('Labels')
    
    plt.tight_layout()
    plt.show()
    
    return dataset_stats

class GroundingDINODetector:
    """Grounding DINO detector for pedestrian behavior detection"""
    
    def __init__(self):
        # Setup device
        if torch.cuda.is_available():
            self.device = "cuda"
            print(f"‚úÖ GPU Available: {torch.cuda.get_device_name(0)}")
            torch.cuda.empty_cache()
        else:
            self.device = "cpu"
            print("‚ö†Ô∏è Using CPU")
        
        # Load Grounding DINO
        print("üîÑ Loading Grounding DINO model...")
        model_id = "IDEA-Research/grounding-dino-tiny"
        self.processor = AutoProcessor.from_pretrained(model_id)
        self.model = AutoModelForZeroShotObjectDetection.from_pretrained(model_id)
        self.model = self.model.to(self.device)
        self.model.eval()
        print("‚úÖ Grounding DINO loaded successfully!")
    
    def detect_objects(self, image, classes=DETECTION_CLASSES, confidence_threshold=CONFIDENCE_THRESHOLD):
        """Detect multiple object classes in image using same logic as original pipeline"""
        # Convert to PIL if needed
        if isinstance(image, np.ndarray):
            if len(image.shape) == 3 and image.shape[2] == 3:
                pil_image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
            else:
                pil_image = Image.fromarray(image)
        else:
            pil_image = image
        
        width, height = pil_image.size
        total_area = width * height
        min_person_area = total_area * (MIN_PERSON_AREA_PERCENTAGE / 100)
        
        # Create text prompt for all classes
        text_prompt = [[class_name for class_name in classes]]
        
        # Prepare inputs
        inputs = self.processor(images=pil_image, text=text_prompt, return_tensors="pt")
        device_inputs = {k: v.to(self.device) if isinstance(v, torch.Tensor) else v 
                        for k, v in inputs.items()}
        
        # Run detection
        with torch.no_grad():
            outputs = self.model(**device_inputs)
        
        # Process results
        target_sizes = torch.tensor([[height, width]]).to(self.device)
        results = self.processor.post_process_grounded_object_detection(
            outputs,
            input_ids=device_inputs.get('input_ids'),
            box_threshold=confidence_threshold,
            text_threshold=0,
            target_sizes=target_sizes
        )[0]
        
        # Extract detections
        raw_detections = []
        if len(results['boxes']) > 0:
            boxes = results['boxes'].cpu().numpy()
            scores = results['scores'].cpu().numpy()
            
            for box, score in zip(boxes, scores):
                x1, y1, x2, y2 = box.astype(int)
                box_area = (x2 - x1) * (y2 - y1)
                
                detection = {
                    'box': [x1, y1, x2, y2],
                    'confidence': float(score),
                    'area': box_area,
                    'area_percentage': (box_area / total_area) * 100
                }
                raw_detections.append(detection)
        
        # Merge overlapping detections using modified logic (person with non-person only)
        merged_detections = merge_person_nonperson_boxes(
            raw_detections, IOU_THRESHOLD, CONTAINMENT_THRESHOLD
        )
        
        # Filter by area
        valid_detections = []
        for det in merged_detections:
            box_area = det['area'] if 'area' in det else (det['box'][2] - det['box'][0]) * (det['box'][3] - det['box'][1])
            area_percentage = (box_area / total_area) * 100
            
            # Update detection with area info
            det['area'] = box_area
            det['area_percentage'] = area_percentage
            
            # Only keep detections with sufficient area
            if area_percentage >= MIN_PERSON_AREA_PERCENTAGE:
                valid_detections.append(det)
        
        return valid_detections, pil_image

class BLIPBARTClassifier:
    """BLIP + BART classification pipeline for behavior analysis"""
    
    def __init__(self):
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        print(f"üîÑ Loading BLIP and BART models on {self.device}...")
        
        # Load BLIP
        blip_model_name = "Salesforce/blip-image-captioning-large"
        self.blip_processor = BlipProcessor.from_pretrained(blip_model_name)
        self.blip_model = BlipForConditionalGeneration.from_pretrained(blip_model_name)
        self.blip_model = self.blip_model.to(self.device)
        self.blip_model.eval()
        
        # Load BART classifier
        self.bart_classifier = pipeline(
            "zero-shot-classification",
            model="facebook/bart-large-mnli",
            device=0 if self.device == "cuda" else -1,
            torch_dtype=torch.float16 if self.device == "cuda" else torch.float32
        )
        
        print("‚úÖ BLIP + BART loaded successfully!")
    
    def preprocess_image(self, image, contrast_factor=1.2, brightness_factor=1.1):
        """Apply preprocessing to image"""
        if isinstance(image, np.ndarray):
            image = Image.fromarray(image)
        
        # Optional enhancement
        enhancer = ImageEnhance.Contrast(image)
        image = enhancer.enhance(contrast_factor)
        enhancer = ImageEnhance.Brightness(image)
        image = enhancer.enhance(brightness_factor)
        
        return image
    
    def generate_caption(self, image, max_length=50):
        """Generate caption using BLIP"""
        try:
            # Apply preprocessing
            processed_image = self.preprocess_image(image)
            
            # Process with BLIP
            inputs = self.blip_processor(processed_image, return_tensors="pt").to(self.device)
            
            with torch.no_grad():
                output = self.blip_model.generate(**inputs, max_length=max_length, num_beams=5)
            
            caption = self.blip_processor.decode(output[0], skip_special_tokens=True)
            return caption
            
        except Exception as e:
            print(f"‚ùå Error generating caption: {e}")
            return "A person in an image."
    
    def classify_behavior(self, caption):
        """Classify behavior using BART"""
        behavior_classes = [
            "walking", "running", "pushing a stroller", 
            "biking", "standing", "skateboarding", "scootering", 
            "person on wheelchair"
        ]
        
        try:
            result = self.bart_classifier(caption, behavior_classes)
            return result['labels'][0], result['scores'][0], result
        except Exception as e:
            print(f"‚ùå Error classifying behavior: {e}")
            return "unknown", 0.0, {"labels": ["unknown"], "scores": [0.0]}

def map_detections_to_annotations(detections, annotations, iou_threshold=IOU_MAPPING_THRESHOLD):
    """Map detected bboxes to ground truth annotations using IoU"""
    mapped_pairs = []
    used_detections = set()
    used_annotations = set()
    
    for det_idx, detection in enumerate(detections):
        best_iou = 0
        best_ann_idx = -1
        
        for ann_idx, annotation in enumerate(annotations):
            if ann_idx in used_annotations:
                continue
                
            iou = calculate_iou(detection['box'], annotation['bbox'])
            if iou >= iou_threshold and iou > best_iou:
                best_iou = iou
                best_ann_idx = ann_idx
        
        if best_ann_idx != -1:
            mapped_pairs.append({
                'detection_idx': det_idx,
                'annotation_idx': best_ann_idx,
                'detection': detection,
                'annotation': annotations[best_ann_idx],
                'iou': best_iou
            })
            used_detections.add(det_idx)
            used_annotations.add(best_ann_idx)
    
    # Find unmatched detections and annotations
    unmatched_detections = [i for i in range(len(detections)) if i not in used_detections]
    unmatched_annotations = [i for i in range(len(annotations)) if i not in used_annotations]
    
    return mapped_pairs, unmatched_detections, unmatched_annotations

def compute_detection_metrics(mapped_pairs, total_detections, total_annotations):
    """Compute detection performance metrics"""
    true_positives = len(mapped_pairs)
    false_positives = total_detections - true_positives
    false_negatives = total_annotations - true_positives
    
    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    return {
        'true_positives': true_positives,
        'false_positives': false_positives,
        'false_negatives': false_negatives,
        'precision': precision,
        'recall': recall,
        'f1_score': f1_score
    }

def display_comprehensive_metrics(all_results):
    """Display comprehensive classification metrics with enhanced analysis"""
    if not all_results:
        print("‚ùå No results to display")
        return None, None
    
    df = pd.DataFrame(all_results)
    
    # Overall accuracy
    overall_accuracy = accuracy_score(df['true_label'], df['predicted_label'])
    print(f"\nüéØ OVERALL CLASSIFICATION ACCURACY: {overall_accuracy:.3f} ({overall_accuracy*100:.1f}%)")
    
    # Get unique labels
    all_labels = sorted(set(df['true_label'].tolist() + df['predicted_label'].tolist()))
    
    # Classification report
    print(f"\nüìä DETAILED CLASSIFICATION REPORT:")
    print("="*80)
    report = classification_report(df['true_label'], df['predicted_label'], 
                                 output_dict=False, zero_division=0)
    print(report)
    
    # Get classification report as dict for further analysis
    report_dict = classification_report(df['true_label'], df['predicted_label'], 
                                      output_dict=True, zero_division=0)
    
    # Per-class metrics with confidence intervals
    precision, recall, f1, support = precision_recall_fscore_support(
        df['true_label'], df['predicted_label'], average=None, zero_division=0, labels=all_labels
    )
    
    metrics_df = pd.DataFrame({
        'Class': all_labels,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Support': support
    })
    
    print(f"\nüìà ENHANCED PER-CLASS METRICS:")
    print("="*80)
    print(metrics_df.round(3).to_string(index=False))
    
    # Class-wise confusion matrices
    cm = confusion_matrix(df['true_label'], df['predicted_label'], labels=all_labels)
    
    # Create comprehensive visualizations
    fig = plt.figure(figsize=(24, 18))
    
    # 1. Overall Confusion Matrix
    plt.subplot(3, 4, 1)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=all_labels, yticklabels=all_labels)
    plt.title('Overall Confusion Matrix', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    
    # 2. Normalized Confusion Matrix
    plt.subplot(3, 4, 2)
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=all_labels, yticklabels=all_labels)
    plt.title('Normalized Confusion Matrix', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    
    # 3. Per-class F1 scores
    plt.subplot(3, 4, 3)
    plt.bar(all_labels, f1, color='lightcoral', alpha=0.7)
    plt.title('F1-Score by Class')
    plt.xlabel('Class')
    plt.ylabel('F1-Score')
    plt.xticks(rotation=45)
    plt.ylim(0, 1)
    for i, v in enumerate(f1):
        plt.text(i, v + 0.01, f'{v:.3f}', ha='center', va='bottom')
    
    # 4. Confidence distribution by correctness
    plt.subplot(3, 4, 4)
    correct_conf = df[df['correct']]['confidence']
    wrong_conf = df[~df['correct']]['confidence']
    
    plt.hist([correct_conf, wrong_conf], bins=20, alpha=0.7, 
             label=['Correct', 'Wrong'], color=['green', 'red'])
    plt.xlabel('Confidence Score')
    plt.ylabel('Frequency')
    plt.title('Confidence Distribution')
    plt.legend()
    
    # 5. IoU distribution
    plt.subplot(3, 4, 5)
    iou_values = df['iou']
    plt.hist(iou_values, bins=20, alpha=0.7, color='blue')
    plt.xlabel('IoU Score')
    plt.ylabel('Frequency')
    plt.title('IoU Distribution')
    plt.axvline(iou_values.mean(), color='red', linestyle='--', 
                label=f'Mean: {iou_values.mean():.3f}')
    plt.legend()
    
    # 6. Class distribution (true vs predicted)
    plt.subplot(3, 4, 6)
    true_counts = df['true_label'].value_counts()
    pred_counts = df['predicted_label'].value_counts()
    
    x = np.arange(len(all_labels))
    width = 0.35
    
    true_vals = [true_counts.get(label, 0) for label in all_labels]
    pred_vals = [pred_counts.get(label, 0) for label in all_labels]
    
    plt.bar(x - width/2, true_vals, width, label='True', alpha=0.7, color='skyblue')
    plt.bar(x + width/2, pred_vals, width, label='Predicted', alpha=0.7, color='orange')
    plt.xlabel('Class')
    plt.ylabel('Count')
    plt.title('Class Distribution: True vs Predicted')
    plt.xticks(x, all_labels, rotation=45)
    plt.legend()
    
    # 7. Precision-Recall by class
    plt.subplot(3, 4, 7)
    plt.scatter(recall, precision, s=100, alpha=0.7, c=range(len(all_labels)), cmap='viridis')
    for i, label in enumerate(all_labels):
        plt.annotate(label, (recall[i], precision[i]), xytext=(5, 5), 
                    textcoords='offset points', fontsize=8)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall by Class')
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.grid(True, alpha=0.3)
    
    # 8. Detection confidence vs IoU
    plt.subplot(3, 4, 8)
    colors = ['green' if correct else 'red' for correct in df['correct']]
    plt.scatter(df['detection_confidence'], df['iou'], c=colors, alpha=0.6)
    plt.xlabel('Detection Confidence')
    plt.ylabel('IoU Score')
    plt.title('Detection Confidence vs IoU')
    plt.grid(True, alpha=0.3)
    
    # 9. Box plot of confidence by class
    plt.subplot(3, 4, 9)
    confidence_by_class = [df[df['true_label'] == label]['confidence'].values for label in all_labels]
    plt.boxplot(confidence_by_class, labels=all_labels)
    plt.xlabel('True Class')
    plt.ylabel('Classification Confidence')
    plt.title('Confidence Distribution by True Class')
    plt.xticks(rotation=45)
    
    # 10. Error analysis by IoU threshold
    plt.subplot(3, 4, 10)
    iou_thresholds = np.arange(0.1, 1.0, 0.1)
    accuracies = []
    for threshold in iou_thresholds:
        high_iou_df = df[df['iou'] >= threshold]
        if len(high_iou_df) > 0:
            acc = accuracy_score(high_iou_df['true_label'], high_iou_df['predicted_label'])
            accuracies.append(acc)
        else:
            accuracies.append(0)
    
    plt.plot(iou_thresholds, accuracies, marker='o', linewidth=2, markersize=6)
    plt.xlabel('IoU Threshold')
    plt.ylabel('Classification Accuracy')
    plt.title('Accuracy vs IoU Threshold')
    plt.grid(True, alpha=0.3)
    
    # 11. Area analysis
    plt.subplot(3, 4, 11)
    area_percentages = df['area_percentage']
    plt.hist(area_percentages, bins=20, alpha=0.7, color='purple')
    plt.xlabel('Area Percentage of Image')
    plt.ylabel('Frequency')
    plt.title('Detection Area Distribution')
    plt.axvline(area_percentages.mean(), color='red', linestyle='--', 
                label=f'Mean: {area_percentages.mean():.1f}%')
    plt.legend()
    
    # 12. Classification errors heatmap
    plt.subplot(3, 4, 12)
    error_matrix = cm - np.diag(np.diag(cm))  # Remove correct predictions
    sns.heatmap(error_matrix, annot=True, fmt='d', cmap='Reds',
                xticklabels=all_labels, yticklabels=all_labels)
    plt.title('Classification Errors Heatmap', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    
    plt.tight_layout()
    plt.show()
    
    # Print detailed error analysis
    print(f"\nüîç DETAILED ERROR ANALYSIS:")
    print("="*80)
    
    # Most confused pairs
    error_pairs = []
    for i, true_label in enumerate(all_labels):
        for j, pred_label in enumerate(all_labels):
            if i != j and cm[i, j] > 0:
                error_pairs.append((true_label, pred_label, cm[i, j]))
    
    error_pairs.sort(key=lambda x: x[2], reverse=True)
    
    print("Top 5 Most Confused Class Pairs:")
    for i, (true_label, pred_label, count) in enumerate(error_pairs[:5]):
        print(f"   {i+1}. {true_label} ‚Üí {pred_label}: {count} errors")
    
    # Best and worst performing classes
    best_class_idx = np.argmax(f1)
    worst_class_idx = np.argmin(f1)
    
    print(f"\nBest Performing Class: {all_labels[best_class_idx]} (F1: {f1[best_class_idx]:.3f})")
    print(f"Worst Performing Class: {all_labels[worst_class_idx]} (F1: {f1[worst_class_idx]:.3f})")
    
    # Confidence statistics
    print(f"\nConfidence Statistics:")
    print(f"   Overall Mean Confidence: {df['confidence'].mean():.3f}")
    print(f"   Correct Predictions Mean Confidence: {correct_conf.mean():.3f}")
    print(f"   Wrong Predictions Mean Confidence: {wrong_conf.mean():.3f}")
    
    # IoU statistics
    print(f"\nIoU Statistics:")
    print(f"   Mean IoU: {df['iou'].mean():.3f}")
    print(f"   IoU Std: {df['iou'].std():.3f}")
    print(f"   Min IoU: {df['iou'].min():.3f}")
    print(f"   Max IoU: {df['iou'].max():.3f}")
    
    return df, metrics_df

def run_pedestrian_behavior_pipeline():
    """Main pipeline for pedestrian behavior dynamics analysis"""
    print("üöÄ PEDESTRIAN BEHAVIOR DYNAMICS PIPELINE")
    print("=" * 80)
    print(f"üéØ Detection Classes: {', '.join(DETECTION_CLASSES)}")
    print(f"üéØ Confidence Threshold: {CONFIDENCE_THRESHOLD}")
    print(f"üîÑ IoU Merge Threshold: {IOU_THRESHOLD}")
    print(f"üìê Containment Threshold: {CONTAINMENT_THRESHOLD}")
    print(f"üìä Min Area Percentage: {MIN_PERSON_AREA_PERCENTAGE}%")
    print(f"üîó IoU Mapping Threshold: {IOU_MAPPING_THRESHOLD}")
    print("=" * 80)
    
    # Setup environment
    setup_environment()
    
    # Create output directory
    os.makedirs(OUTPUT_FOLDER, exist_ok=True)
    
    # Find image files
    image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
    image_files = []
    
    # Check if folders exist
    if not os.path.exists(IMAGES_FOLDER):
        print(f"‚ùå Images folder not found: {IMAGES_FOLDER}")
        return None
    
    if not os.path.exists(ANNOTATIONS_FOLDER):
        print(f"‚ùå Annotations folder not found: {ANNOTATIONS_FOLDER}")
        return None
    
    for file in os.listdir(IMAGES_FOLDER):
        if any(file.lower().endswith(ext) for ext in image_extensions):
            image_files.append(file)
    
    if not image_files:
        print(f"‚ùå No image files found in {IMAGES_FOLDER}")
        return None
    
    print(f"üñºÔ∏è Found {len(image_files)} images")
    
    # Analyze dataset statistics
    dataset_stats = analyze_dataset_statistics(image_files, IMAGES_FOLDER, ANNOTATIONS_FOLDER)
    
    # Initialize models
    detector = GroundingDINODetector()
    classifier = BLIPBARTClassifier()
    
    # Pipeline results storage
    all_results = []
    detection_metrics = []
    all_captions = []
    
    print(f"\nüîÑ Processing {len(image_files)} images...")
    
    for i, image_file in enumerate(tqdm(image_files, desc="Processing images")):
        image_path = os.path.join(IMAGES_FOLDER, image_file)
        image_id = os.path.splitext(image_file)[0]
        xml_path = os.path.join(ANNOTATIONS_FOLDER, f"{image_id}.xml")
        
        print(f"\n{'='*60}")
        print(f"üì∑ IMAGE {i+1}/{len(image_files)}: {image_file}")
        
        if not os.path.exists(xml_path):
            print(f"‚ùå No annotation found for {image_file}")
            continue
        
        # Load image and annotation
        image = cv2.imread(image_path)
        if image is None:
            print(f"‚ùå Could not load image: {image_file}")
            continue
        
        annotations, image_info = parse_xml_annotation(xml_path)
        if not annotations:
            print(f"‚ùå No valid annotations found in {xml_path}")
            continue
        
        print(f"üìã Ground truth annotations: {len(annotations)}")
        for j, ann in enumerate(annotations):
            print(f"   GT {j+1}: {ann['label']} at {ann['bbox']}")
        
        # Run Grounding DINO detection
        detections, pil_image = detector.detect_objects(image)
        print(f"üîç Grounding DINO detections: {len(detections)}")
        
        # Map detections to annotations
        mapped_pairs, unmatched_detections, unmatched_annotations = map_detections_to_annotations(
            detections, annotations, IOU_MAPPING_THRESHOLD
        )
        
        print(f"üîó Mapping results:")
        print(f"   ‚úÖ Mapped pairs: {len(mapped_pairs)}")
        print(f"   üì¶ Unmatched detections: {len(unmatched_detections)} (discarded)")
        print(f"   üìã Unmatched annotations: {len(unmatched_annotations)}")
        
        # Compute detection metrics
        det_metrics = compute_detection_metrics(mapped_pairs, len(detections), len(annotations))
        det_metrics['image_file'] = image_file
        detection_metrics.append(det_metrics)
        
        print(f"üìä Detection Metrics - P: {det_metrics['precision']:.3f}, R: {det_metrics['recall']:.3f}, F1: {det_metrics['f1_score']:.3f}")
        
        # Process only mapped pairs for classification
        print(f"üß† Running BLIP + BART classification on {len(mapped_pairs)} mapped detections...")
        
        for pair_idx, pair in enumerate(mapped_pairs):
            detection = pair['detection']
            annotation = pair['annotation']
            x1, y1, x2, y2 = detection['box']
            
            # Crop detection
            img_array = np.array(pil_image)
            crop = img_array[y1:y2, x1:x2]
            
            if crop.size > 0:
                crop_pil = Image.fromarray(crop)
                
                # Generate caption with BLIP
                caption = classifier.generate_caption(crop_pil)
                all_captions.append(caption)
                
                # Classify behavior with BART
                pred_label, pred_conf, all_scores = classifier.classify_behavior(caption)
                
                # Store comprehensive results
                result = {
                    'image_file': image_file,
                    'detection_idx': pair['detection_idx'],
                    'annotation_idx': pair['annotation_idx'],
                    'true_label': annotation['label'],
                    'predicted_label': pred_label,
                    'confidence': pred_conf,
                    'blip_caption': caption,
                    'iou': pair['iou'],
                    'detection_bbox': detection['box'],
                    'annotation_bbox': annotation['bbox'],
                    'detection_confidence': detection['confidence'],
                    'detection_area': detection['area'],
                    'area_percentage': detection['area_percentage'],
                    'correct': pred_label == annotation['label']
                }
                all_results.append(result)
                
                # Print caption and classification result
                print(f"   üéØ Crop {pair_idx+1}:")
                print(f"      üìù Caption: '{caption}'")
                print(f"      üè∑Ô∏è  Classification: {annotation['label']} -> {pred_label} ({pred_conf:.3f}) | IoU: {pair['iou']:.3f}")
        
        # Clean GPU memory
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
    
    # Save detailed results
    if all_results:
        results_df = pd.DataFrame(all_results)
        results_df.to_csv(os.path.join(OUTPUT_FOLDER, 'pedestrian_behavior_results.csv'), index=False)
        
        detection_df = pd.DataFrame(detection_metrics)
        detection_df.to_csv(os.path.join(OUTPUT_FOLDER, 'detection_metrics.csv'), index=False)
        
        # Save captions
        captions_df = pd.DataFrame({'captions': all_captions})
        captions_df.to_csv(os.path.join(OUTPUT_FOLDER, 'generated_captions.csv'), index=False)
        
        # Display comprehensive metrics
        print(f"\nüéâ PIPELINE COMPLETE!")
        print(f"üíæ Results saved to: {OUTPUT_FOLDER}")
        print(f"üìä Total processed detections: {len(all_results)}")
        print(f"üìù Total generated captions: {len(all_captions)}")
        
        # Detection Performance Summary
        if detection_metrics:
            avg_precision = np.mean([m['precision'] for m in detection_metrics])
            avg_recall = np.mean([m['recall'] for m in detection_metrics])
            avg_f1 = np.mean([m['f1_score'] for m in detection_metrics])
            
            print(f"\nüîç DETECTION PERFORMANCE SUMMARY:")
            print(f"   üìà Average Precision: {avg_precision:.3f}")
            print(f"   üìà Average Recall: {avg_recall:.3f}")
            print(f"   üìà Average F1-Score: {avg_f1:.3f}")
        
        # Classification Performance with enhanced metrics
        final_df, metrics_df = display_comprehensive_metrics(all_results)
        
        # Display sample captions
        print(f"\nüìù SAMPLE GENERATED CAPTIONS:")
        print("="*60)
        for i, caption in enumerate(all_captions[:10]):
            print(f"   {i+1}. {caption}")
        if len(all_captions) > 10:
            print(f"   ... and {len(all_captions)-10} more captions")
        
        return final_df, detection_df, metrics_df, dataset_stats
    
    else:
        print("‚ùå No results to display")
        return None, None, None, None

# Run the complete pipeline
print("üöÄ Starting Pedestrian Behavior Dynamics Pipeline...")
results_df, detection_df, class_metrics_df, dataset_stats = run_pedestrian_behavior_pipeline()
```
